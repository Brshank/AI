### 1. Generate a sinusoidal signal and plot it

```python
import numpy as np
import matplotlib.pyplot as plt

```

# >TASK 1<

```python
x=np.linspace(0,4*np.pi,2048)
y=np.sin(x)

plt.plot(x,y)
plt.xlabel("Time (t) ---->")
plt.ylabel(" Sin(t) ---->")
plt.title("1 Sinusodial Signal ")
```

# >TASK 2<

```python
y.shape
```

```python
r=np.correlate(y,y,mode='full')

m=np.linspace(-4*np.pi,+4*np.pi,2*len(y)-1)

plt.plot(m,r)
plt.xlabel("Time Lag Correlation Factor---->")
plt.ylabel("Magnitude---->")
plt.title('Sinusodial Signal for Correlation')
```

```python
r[2048]
```

## Analysis of the properties of autocorrelation:

### 2.Rxx is Maximum at tau=0
### 3.The mean square value is 1023.48

# >Task 3<

## 3. Add a noise signal with original signal and plot it

```python
noise = np.random.normal(1, 1, 2048)
yn = y + noise

plt.plot(x,yn)
plt.ylabel("sin(t)")
plt.xlabel("time 't'")
plt.title('Noisy Sinusoidal signal')
```

# >Task 4<

### 4. Find the cross-correlation function of the signal with noisy signal and verify the properties of crosscorrelation using plots

```python
rn = np.correlate(y, yn, mode = 'full')
m = np.linspace(-4*np.pi, 4*np.pi, 2 * len(y) - 1)
plt.plot(m, rn)
plt.ylabel("magnitude")
plt.xlabel("time lag ")
plt.title('Sinusoidal signal')
```

```python
nn = np.correlate(noise, noise, mode = 'full')  #noise
ry = np.correlate(y, y, mode = 'full')  #original
plt.plot(nn)
plt.plot(ry)
plt.plot(rn)
plt.legend(nn)
```
