## 3. <Task 3> WAP to find PSD of a sine signal

```python
F1= 200 # freq1
F2= 400# freq2

fs=2000   #sampling freq
T=5    # samples in how manyseconds 
t=np.arange(0,T,(1/fs))   #time instances
```

```python
plt.figure(figsize=(15,3))
x1=np.sin(2*np.pi*F1*t)
plt.plot(t,x1)
```

```python
plt.figure(figsize=(15,3))
x2=np.sin(2*np.pi*F2*t)
plt.plot(t,x2)
```

```python
plt.figure(figsize=(15,3))
xn=x1+2*x2  #amplitude twice
plt.plot(t,xn)
```

```python
xn.shape    #fs*T
```

```python
from IPython.display import Audio
Audio(data=xn,rate=fs)
```

```python
Xk=np.fft.fft(xn)
PSD=np.abs(Xk)
N = len(xn)
k = np.arange(N) * (fs/N)
```

```python
plt.figure(figsize=(15,3))
plt.stem(k[0:int(N/2)], PSD[0:int(N/2)])
plt.xlabel('Frequency')
plt.ylabel('X(k)')
plt.title('PSD')
plt.show()
```

# Task 4 WAP atp calculate PSD of a noise/random signal

```python
n=np.random.normal(-1,+1,len(xn))
```

```python
n.shape
```

```python
yn = xn + n
```

```python
plt.figure(figsize=(15,3))
plt.plot(t, yn)
plt.show()
```

```python
ryn = np.correlate(yn, yn, mode='same')
```

```python
plt.figure(figsize=(15,3))
plt.plot(ryn)
plt.show()
```

```python
Xk = np.fft.fft(ryn)
PSD = np.abs(Xk)
```

```python
k = np.arange(N) * (fs/N)
```

```python
plt.figure(figsize=(15,3))
plt.stem(k[0:int(N/2)], PSD[0:int(N/2)])
plt.xlabel('Frequency')
plt.ylabel('X(k)')
plt.title('PSD')
plt.show()
```
