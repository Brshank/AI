## 1. Write a program to calculate thge fourier transform

```python
fs=1000
t=np.arange(1000)/fs
signala=np.sin(2*np.pi*100*t)
plt.figure(figsize=(14,4))
plt.plot(t,signala)
plt.xlabel("time-->")
plt.ylabel('magnitude')
```

```python
f2=abs(np.fft.fft(signala))
plt.stem(f2[0:500])
plt.xlabel("frequencey-->")
plt.ylabel('magnitude')
```

```python
signalb=np.sin(2*np.pi*20*t)
plt.figure(figsize= (14,4))
plt.plot(t,signalb)
plt.xlabel("time-->")
plt.ylabel('magnitude')
```

```python
f2=abs(np.fft.fft(signalb))
plt.stem(f2[0:100])
plt.xlabel("frequencey-->")
plt.ylabel('magnitude')
```

```python
fs, data = wavfile.read('Forest.wav')
```

```python
from IPython.display import Audio
Audio(data=data,rate=fs)
```

```python
plt.figure(figsize=(14,4))
plt.plot(data)
```

```python
plt.figure(figsize=(14,4))
f4=abs(np.fft.fft(data))
plt.plot(f4[0:int(len(f4)/2)])
```

```python
fc=1000
w=fc/(fs/2)
b,a = signal.butter(5,w,'high')
output = signal.filtfilt(b, a,data)
```

```python
Audio(data=output,rate=fs)
```

```python
plt.figure(figsize=(14,4))
f5=abs(np.fft.fft(output))
plt.plot(f5[0:int(len(f5)/2)])
```
