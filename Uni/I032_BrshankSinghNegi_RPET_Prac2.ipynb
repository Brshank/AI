```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import warnings
warnings.filterwarnings('ignore')
```

## 1. Write a program to generate PDF of uniform distribution function.
## $$ f(x)=\begin{cases} \frac{1}{b-a} & \text{ ; } a\leq x\leq b \\ 0 & \text{ ; } otherwise \end{cases} $$


```python
a=10
b=50
fx_uniform= scipy.stats.uniform.rvs(loc=a,scale=b-a,size=20000)

bx = sns.distplot(fx_uniform)
bx.set(xlabel = 'x-->' , ylabel = 'Frequency' , title='Uniform Distribution')
```

## 2.Write a program to generate PDF of Normal distribution function.
## $$ {f(x)=\frac{1}{2 \pi \sigma^{2}}.e^{\frac{-(x-\mu)^{2}}{2\sigma^{2}}}} $$

```python
n = 10000
mean = 0 
std = 1

fx_normal = scipy.stats.norm.rvs(size=n , loc=mean , scale=std )


bx = sns.distplot(fx_normal)
bx.set(xlabel = 'x-->' , ylabel = 'Frequency' , title='Normal Distribution')
```

## 3.Write a program to generate PDF of Gamma distribution function.
## $$f(x)=\frac{\beta^{\alpha}\; x^{\alpha-1}\;e^{-\beta x} }{\Gamma (a)}$$

```python
n = 10000
mean = 0 
std = 1

fx_gamma = scipy.stats.gamma.rvs(size=n , loc=mean , scale=std ,a=5)


bx = sns.distplot(fx_gamma)
bx.set(xlabel = 'x-->' , ylabel = 'Frequency' , title='Gamma Distribution')
```

## 4.Write a program to generate PDF of Exponential distribution function.
## $$ f(x)=\begin{cases}\lambda e^{-\lambda x} & \text{ if } x\geq 0 \\ 0 & \text { if } x<0 \end{cases} $$

```python
n = 10000
mean = 0 
std = 1

fx_expon = scipy.stats.expon.rvs(size=n , loc=mean , scale=std)


bx = sns.distplot(fx_expon)
bx.set(xlabel = 'x-->' , ylabel = 'Frequency' , title='Exponential Distribution')
```

## 5.Write a program to generate PDF of Binomial distribution function.
## $$ f(x)=_{x}^{n}\textrm{C}\;\; p^{x}\;\; (1-p)^{n-x} $$

```python
m = 10000
fx_binom = scipy.stats.binom.rvs(size = m , n = 10 , p = 0.8)

bx = sns.distplot(fx_binom)
bx.set(xlabel = 'x-->' , ylabel = 'Frequency' , title='Binomial Distribution')
```

## 6.Write a program to generate PDF of Poisson’s distribution function.
## $$P(x)=\frac{e^{-\lambda}\;\;\lambda^{x}}{x!}$$

```python
m = 10000
fx_poison = scipy.stats.poisson.rvs(size = m , mu=3)

bx = sns.distplot(fx_poison,)
bx.set(xlabel = 'x-->' , ylabel = 'Frequency' , title='Poisson Distribution')
```
